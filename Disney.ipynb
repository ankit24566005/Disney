## *Introduction to the dataset*  

```python
# Importing Necessary Libraries and Loading the Dataset
import pandas as pd
df=pd.read_csv('disney_plus_titles.csv')
df.head()
```

```python
# Describe the dataset
df.info()
df.describe()
df.isnull().sum()
```

# *Univariate Analysis*

```python
# Univariate Analysis on  Numerical Variables

# Now check for any NaN values that were created during the conversion
df['release_year'].isna().sum()

# Attempt to convert problematic columns to numeric
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

a1 = df['release_year'].mean()
a2 = df['release_year'].median()
a3 = df['release_year'].std()
print(a1)
print(a2)
print(a3)
```

```python
# Plotting Distribution of Numerical Variables

import matplotlib.pyplot as plt
df['release_year'].hist(bins=20)
plt.show()
```

Categorical Data Analysis

```python
df['type'].value_counts().plot(kind='bar')
plt.show()
```

# *Multivariate Analysis*

```python
import seaborn as sns
numeric_df=df.select_dtypes(include=['float64', 'int64'])
correlation_matrix=numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# bar plot

pd.crosstab(df['type'], df['release_year']).plot(kind='bar')
plt.show()
```

```python
# Scatter plot

sns.scatterplot(x='release_year',y='release_year', hue='type', data=df)
plt.show()
```
